```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
n = np.arange(1, 100002)
probability = 1 - (1 - (1/n))**n
fig, ax = plt.subplots(figsize = (9, 7))
ax.plot(n, probability)
ax.axhline(1 - 1/np.e, linestyle = ":", linewidth = 5, color = "red", label = "1 - 1/e")
ax.legend(loc = "best")
ax.set(xscale = "log", xlabel = "Number of observations, n", 
       ylabel = "Probability jth observation is in the bootstrap ");
```
